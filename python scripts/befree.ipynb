```python
import requests
import pandas as pd
from io import StringIO

url = "https://maayanlab.cloud/Harmonizome/api/1.0/dataset/DisGeNET+Gene-Disease+Associations"
response = requests.get(url)
response.raise_for_status()

data = response.json()
columns = ["Gene", "Disease", "Association Type", "Evidence", "Source"]
df = pd.DataFrame(data, columns=columns)

print(df.head())
```
